Aggregating multiple columns with a groupby

```python
was_2020.groupby("State").agg(
    max_methane=pd.NamedAgg(column='CH4 (ppmv height)', aggfunc='max'), 
    mean_dms=pd.NamedAgg(column="DMS (MS)", aggfunc='mean')
)
```

```python
was_2020.groupby(["State", "Wind Spd (mph)"]).agg(
    max_methane=pd.NamedAgg(column='CH4 (ppmv height)', aggfunc='max'), 
    mean_dms=pd.NamedAgg(column="DMS (MS)", aggfunc='mean')
)
```

```python
# Set new value
was_2020.at[141, 'Altitude (m)'] = 35
```

```python
was_2020.iloc[141]['Altitude (m)']
```

```python
# Standardize the datatype from object to float https://stackoverflow.com/questions/37561991/what-is-dtypeo-in-pandas
was_2020['Altitude (m)'] = pd.to_numeric(was_2020['Altitude (m)'])
```

```python
# Confirm it worked
was_2020['Altitude (m)'].dtype
# ðŸŽ‰
```

Binning

```python
# Letting pandas do your sampling
# bins = pd.cut(was_2020["Altitude (m)"], bins=3, labels=("low", "mid", "high"), retbins=True)
```

```python
was_2020['Altitude (m)'].describe()
```

```python
# Specify your bin values
bins = pd.IntervalIndex.from_tuples([(-50, 100), (100, 800), (800, 2500)])
```

```python
bins
```

```python
# Cut the Altitude values based on the bins
bins = pd.cut(was_2020['Altitude (m)'], bins)
```

```python
was_2020.groupby(bins)['CO2 (ppmv)'].mean()
```

```python
was_2020.groupby(['State', bins]).agg(
    max_methane=pd.NamedAgg(column='CH4 (ppmv height)', aggfunc='max'), 
    mean_dms=pd.NamedAgg(column="DMS (MS)", aggfunc='mean')
)
```
