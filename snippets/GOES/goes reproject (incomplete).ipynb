# Assigning a CRS

#### `cartopy`

```python
import cartopy
```

```python
goes_crs = cartopy.crs.Geostationary(central_longitude=-75, sweep_axis='x')
```

#### from proj4

```python
from rasterio.crs import CRS
```

```python
goes_crs = CRS.from_proj4('+proj=geos +h=35785831.0 +lon_0=-75 +sweep=x')
```

### Reproject a vector point

```python
from pyproj import Transformer
from shapely.geometry import Point
from shapely.ops import transform
```

```python
# create the "Transformer object" -- the thing that we will use to convert our points
# INPUTS: start_proj, end_proj
t = Transformer.from_crs('epsg:4326', goes_crs, always_xy=True).transform
```

```python
point_goes = transform(t, Point(-119.5, 34.5))
```

```python
list(point_goes.coords)
```
