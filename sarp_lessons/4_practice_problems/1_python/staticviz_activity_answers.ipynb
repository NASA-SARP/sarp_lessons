# Altitude Plot Visualization Activity

1. Start with that naames dataset that we were working with in lecture. The data file is called `naames-mrg01-c130_merge_20151112_R5_thru20151114.csv` and it is in the tabular dataâ€™s data folder. Create a new notebook for this activity and use pandas to open the data file.

```python
import pandas as pd
```

```python
# Import example data - flight from NAAMES campaign
naames = pd.read_csv('../data/naames-mrg01-c130_merge_20151112_R5_thru20151114.csv', skiprows=223)
```

2. This dataset has a nodata value of -999999. Replace all -999999 values with `np.nan`. 

```python
import numpy as np
```

```python
naames = naames.replace(-999999, np.nan)
```

3. Make a plot of the full dataset with Fractional Day on the x axis and Altitude on the y axis

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.scatter(naames[' Fractional_Day'], naames[' ALTP'])
# ax.plot(naames[' Fractional_Day'], naames[' ALTP']) would also work at this stage, although
# you will have to switch to ax.scatter when you add Methanol to the color scale.
```

4. It looks like we have two days of data here - julian day 316 and julian day 318. Filter the data so that we are only looking at Julian day 316. Re-run the plot with the subset of data.

```python
naames = naames[naames[' Fractional_Day'] < 317]
```

```python
fig, ax = plt.subplots()

ax.scatter(naames[' Fractional_Day'], naames[' ALTP'])
```

5. Let's add some labels! Label both the x and y axes and give the plot a title.

```python
fig, ax = plt.subplots()

ax.scatter(naames[' Fractional_Day'], naames[' ALTP'])

ax.set_ylabel('Altitude (km)')

ax.set_title('$H_{2}O\:and\:Methanol\:by\:Altitude\:over\:Time$')
ax.set_xlabel('Julian Day')
```

6. A very common approach is to color an altitude profile with the values of a particular pollutant. Color the altitude line by the column ` Methanol_MixingRatio`.

```python
fig, ax = plt.subplots()

ax.scatter(naames[' Fractional_Day'], naames[' ALTP'], c=naames[' Methanol_MixingRatio'])

ax.set_ylabel('Altitude (km)')

ax.set_title('$H_{2}O\:and\:Methanol\:by\:Altitude\:over\:Time$')
ax.set_xlabel('Julian Day')
```

7. Now let's add the `H2O_MixingRatio_LARGE` mixing ratio to the plot. It can be added as a seperate line. Set the color to black.

```python
fig, ax = plt.subplots()

ax.scatter(naames[' Fractional_Day'], naames[' ALTP'], c=naames[' Methanol_MixingRatio'])

ax.plot(naames[' Fractional_Day'], naames[' H2O_MixingRatio_LARGE'], color='black')

ax.set_ylabel('Altitude (km)')

ax.set_title('$H_{2}O\:and\:Methanol\:by\:Altitude\:over\:Time$')
ax.set_xlabel('Julian Day')
```

8. We have two different variables on our plot with two very different scales (You could confirm that by running `.describe()` on each of the two data columns). To improve our plot, change the axes so that the altitude data is using one scale and the H2O mixing ratio is using a different scale. Give the second axis its own label.

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.scatter(naames[' Fractional_Day'], naames[' ALTP'], c=naames[' Methanol_MixingRatio'])

ax2.plot(naames[' Fractional_Day'], naames[' H2O_MixingRatio_LARGE'], color='black')
ax2.set_ylabel('$H_{2}O\:Mixing\:Ratio$')

ax1.set_ylabel('Altitude (km)')

ax1.set_title('$H_{2}O\:and\:Methanol\:by\:Altitude\:over\:Time$')
ax1.set_xlabel('Julian Day')
```

9. The data feels a little squished. Let's adjust the size of the plot to be a longer rectangle so we can see better.

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

fig.set_size_inches(15, 5)

ax1.scatter(naames[' Fractional_Day'], naames[' ALTP'], c=naames[' Methanol_MixingRatio'])

ax2.plot(naames[' Fractional_Day'], naames[' H2O_MixingRatio_LARGE'], color='black')
ax2.set_ylabel('$H_{2}O\:Mixing\:Ratio$')

ax1.set_ylabel('Altitude (km)')

ax1.set_title('$H_{2}O\:and\:Methanol\:by\:Altitude\:over\:Time$')
ax1.set_xlabel('Julian Day')
```

10. To get a better idea of what the methanol data is telling us, add a colorbar. If you get a colorbar but it overlaps the axis data, first celebrate (that's great progress). Then adjust the colorbar so that it doesn't overlap the axis.

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

fig.set_size_inches(15, 5)

scatter1 = ax1.scatter(naames[' Fractional_Day'], naames[' ALTP'], c=naames[' Methanol_MixingRatio'])

ax2.plot(naames[' Fractional_Day'], naames[' H2O_MixingRatio_LARGE'], color='black')
ax2.set_ylabel('$H_{2}O\:Mixing\:Ratio$')

ax1.set_ylabel('Altitude (km)')

ax1.set_title('$H_{2}O\:and\:Methanol\:by\:Altitude\:over\:Time$')
ax1.set_xlabel('Julian Day')

fig.colorbar(scatter1, ax=ax1, label="Methanol Mixing Ratio", pad=0.08) 
# orientation = horizontal is also an option to place the colorbar below the data
```
