## Dimensions

Below are 3 different xarray DataArrays. Inspect the HTML representation of each array and draw each DataArray on a piece of paper.

```python
ndvi_values_4 = np.random.rand(4, 5, 6, 2).round(2)
```

### DataArray no. 1

```python
ndvi_values_2 = ndvi_values_4[0, :, :, 0]

latitude  = [34.3, 34.4, 34.5, 34.6, 34.7]
longitude = [-118.3, -118.2, -118.1, -118.0, -117.9, -117.8]

ndvi_no1 = xr.DataArray(
    ndvi_values_2, 
    dims = ['latitude', 'longitude'],
    coords=[latitude, longitude]
    )
```

```python
ndvi_no1
```

### DataArray no. 2

```python
ndvi_values_3 = ndvi_values_4[:, :, :, 0]

time = ['2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01']
latitude  = [34.3, 34.4, 34.5, 34.6, 34.7]
longitude = [-118.3, -118.2, -118.1, -118.0, -117.9, -117.8]

ndvi_no2 = xr.DataArray(
    ndvi_values_3, 
    dims = ['time', 'latitude', 'longitude'],
    coords=[time, latitude, longitude]
    )
```

### DataArray no. 3
