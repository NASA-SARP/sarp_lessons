```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import example data - flight from NAAMES campaign
naames = pd.read_csv('../data/naames-mrg01-c130_merge_20151112_R5_thru20151123.csv', skiprows=223)
# Filter down to just 1 day
naames = naames[naames[' Fractional_Day'] < 317]
# Remove NaN values
naames = naames.replace({-999999: np.nan})
```

1. Below is the one of the first plots we created in the lesson. What is the class of the the input arguments?

`plt.plot(naames[' Fractional_Day'], naames[' ALTP'])`

_They are both a pandas Series_. You can see this using `naames[' Fractional_Day'].__class__`.
